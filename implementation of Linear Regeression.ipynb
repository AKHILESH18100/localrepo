# Simple Linear regeression

```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create Dataset
X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([45000,50000,60000,65000,70000])

# Step 3: Create Model
model = LinearRegression()

# Step 4: Train Model
model.fit(X,Y)

# Step 5: Show Regression Equation
print("Intercept (a): ", model.intercept_)
print("Slope (b): ", model.coef_[0])
print("Regression Equation: Y =", model.intercept_, "+", model.coef_[0], "* X")

# Step 6: Plot Graph
Y_pred = model.predict(X)
plt.scatter(X,Y)
plt.plot(X,Y_pred)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Implementation")
plt.show()

# Step 7: Take Input from User
exp = float(input("Enter Years of Experience to Predict Salary: "))

# Convert input into 2D array
exp = np.array([[exp]])

# Step 8: Predict Salary
predicted_salary = model.predict(exp)

print("Predicted Salary is:", predicted_salary[0])
```

# Multiple Linear Regeression

```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# Step 2: Dataset
X = np.array([
    [1,12],
    [2,14],
    [3,16],
    [4,18],
    [5,20]
])

Y = np.array([40000,50000,60000,70000,80000])

# Step 3: Train Model
model = LinearRegression()
model.fit(X,Y)

# Step 4: Regression Equation
print("Intercept (a):", model.intercept_)
print("Coefficients (b1, b2):", model.coef_)

print("Regression Equation:")
print("Y =", model.intercept_,
      "+", model.coef_[0], "* X1 +",
      model.coef_[1], "* X2")

# Step 5: GRAPH PART
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Step 6: Scatter plot (actual data)
ax.scatter(X[:,0], X[:,1], Y)

# Step 7: Create meshgrid for plane
x1_surf, x2_surf = np.meshgrid(
    np.linspace(X[:,0].min(), X[:,0].max(), 10),
    np.linspace(X[:,1].min(), X[:,1].max(), 10)
)

y_surf = model.intercept_ + \
         model.coef_[0]*x1_surf + \
         model.coef_[1]*x2_surf

# Step 8: Plot regression plane
ax.plot_surface(x1_surf, x2_surf, y_surf)

ax.set_xlabel("Experience")
ax.set_ylabel("Education")
ax.set_zlabel("Salary")
ax.set_title("Multiple Linear Regression Graph")

plt.show()

# Step 9: User Input
exp = float(input("Enter Years of Experience: "))
edu = float(input("Enter Education Level: "))

new_data = np.array([[exp, edu]])
prediction = model.predict(new_data)

print("Predicted Salary:", prediction[0])

# Step 10: Predict values for training data
Y_pred = model.predict(X)

# Step: Calculate Residuals
residuals = Y - Y_pred

# Step: Residual Plot
plt.figure()
plt.scatter(Y_pred, residuals)

plt.axhline(y=0)

plt.xlabel("Predicted Salary")
plt.ylabel("Residual Error")
plt.title("Residual Error Plot")

plt.show()
```
